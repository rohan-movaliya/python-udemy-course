```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data
data = {
    "February": range(1, 27),
    "March": range(27, 31),
    "April": range(31, 61),
    "May": range(61, 73),
    "June": range(73, 101)
}

# Flatten the data for easier manipulation
days = []
months = []
for month, day_range in data.items():
    days.extend(day_range)
    months.extend([month] * len(day_range))

df = pd.DataFrame({"Day": days, "Month": months})


# Count the number of days in each month
month_counts = df["Month"].value_counts().sort_index()


# Pie Chart
plt.figure(figsize=(8, 8))
month_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Proportion of Days Spent in Each Month')
plt.ylabel('')
plt.show()

```
